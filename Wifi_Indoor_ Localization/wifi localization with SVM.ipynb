Classifier selection

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=2000, criterion='gini', max_depth=4,min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None,bootstrap=True, oob_score=False, n_jobs=10, random_state=0, verbose=0, warm_start=False,class_weight=None)
```

```python

data=wifi.loc[:,['wifi_signal_3','wifi_signal_5']]
```

```python

labels=wifi.loc[:,['Room_number']].values
```

```python
data= data.transform(lambda x: (x - x.mean()) / x.std())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import learning_curve
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.4, random_state=42)
```

```python
clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
title = 'Learning Curves (Random Forest)'
estimator = clf
plot_learning_curve(estimator, title, X_train, y_train)
plt.show()
```

```python
import tensorflow as tf
import sys
from keras.models import Sequential
model = Sequential()
from keras.layers import Dense
model.add(Dense(units=3000, activation='relu', input_dim=2))
model.add(Dense(units=2500, activation='relu'))
model.add(Dense(units=2000, activation='relu'))
model.add(Dense(units=1500, activation='relu'))
model.add(Dense(units=1000, activation='relu'))



model.add(Dense(units=1, activation='softmax'))
```

```python
model.compile(loss='mse',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=1600)
```
